```python
for filename in os.listdir('/Users/amyel/research/vmc/allfits/'):
    
    print(filename)
    
    fx = fitsio.FITS('/Users/amyel/research/vmc/allfits/'+filename)

    if filename[3] == '_':
        print(filename[:3])
        tiles[filename[:3]]=fx[1].read()
    else:
        print(filename[:4])
        tiles[filename[:4]]=fx[1].read()
```
