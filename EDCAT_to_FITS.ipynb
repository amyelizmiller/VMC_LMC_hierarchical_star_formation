```python
tiles={}
tilenames=[]
```

```python
for filename in os.listdir('/Users/amyel/research/vmc/ECDAT/'):
    print(filename)
    newstr = ''.join((ch if ch in '0123456789_' else ' ') for ch in filename)
    print(newstr[4:8])
    
    tilenames.append(newstr[4:8])
    tiles[newstr[4:8]]=np.genfromtxt('/Users/amyel/research/vmc/ECDAT/'+str(filename),deletechars='#',skip_header =28,names=True)
            
```

```python
names=['J_MAG','J_ERR','Ks_MAG','Ks_ERR','AVERAGE_RA','AVERAGE_DEC','Lcomp_J','Lcomp_Ks','J_SHARP','Ks_SHARP','syserr_J','syserr_Ks','Star_prob','RA_Y','DEC_Y','Y_MAG','Y_ERR','Y_SHARP','RA_J','DEC_J','RA_Ks','DEC_Ks','Lcomp_Y','syserr_Y','nY,nJ,nKs']
```

```python
names
```

```python
stars={}
```

```python
def mag_cut(tile):
    print(tile)
    
    stars[tile]={}
    
    selected = np.where((tiles[tile]['Y_MAG']>=0)&(tiles[tile]['Y_MAG']<=30))
    
    for name in names: 
        print(name)
        stars[tile][name]=tiles[tile][name][selected]
```

```python
for key in tiles.keys():
    mag_cut(key)
```

```python
tiles.keys()
```

```python
fig, ax = plt.subplots(figsize=(30,20))
for tile in stars.keys():
    
    ax.scatter(stars[tile]['RA_Y'],stars[tile]['DEC_Y'],s=0.1)
ax.set_xlim(100,60)
ax.set_ylim(-80,-60)
```

```python
def save(tile):
    
    print(tile)
    
    array_list = [tiles[tile][name] for name in names]
    
    fitsio.write('/Users/amyel/research/vmc/allfits/'+tile+'data.fits',array_list,names=names)
    
    
    return 0
```

```python
for tile in tiles.keys():
    print(tile)
    save(tile)
```
